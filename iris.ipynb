```python
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df['variety'].unique()
```

```python
Setosa = df[df['variety']=='Setosa']
Versicolor = df[df['variety']=='Versicolor']
Virginica = df[df['variety']=='Virginica']
```

```python
import plotly.offline as py
import plotly.graph_objs as go
from plotly import tools
py.init_notebook_mode(connected=True)
```

```python
df.columns
```

```python
trace1 = go.Scatter(
    x = Setosa['sepal.length'],
    y = Setosa['sepal.width'],
    mode = 'markers',
    name = 'Setosa')
trace2 = go.Scatter(
    x = Versicolor['sepal.length'],
    y = Versicolor['sepal.width'],
    mode = 'markers',
    name = 'Versicolor')
trace3 = go.Scatter(
    x = Virginica['sepal.length'],
    y = Virginica['sepal.width'],
    mode = 'markers',
    name = 'Virginica')

trace4 = go.Scatter(
    x = Setosa['petal.length'],
    y = Setosa['petal.width'],
    mode = 'markers',
    name = 'Setosa')
trace5 = go.Scatter(
    x = Versicolor['petal.length'],
    y = Versicolor['petal.width'],
    mode = 'markers',
    name = 'Versicolor')
trace6 = go.Scatter(
    x = Virginica['petal.length'],
    y = Virginica['petal.width'],
    mode = 'markers',
    name = 'Virginica')


fig = tools.make_subplots(rows=1,cols=2,subplot_titles=['sepal','petal'])
fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,1)
fig.append_trace(trace3,1,1)
fig.append_trace(trace4,1,2)
fig.append_trace(trace5,1,2)
fig.append_trace(trace6,1,2)
fig['layout'].update(
    xaxis1 = dict(title = 'length'),
    yaxis1 = dict(title = 'width'),
    xaxis2 = dict(title = 'length'),
    yaxis2 = dict(title = 'width'),    
)
py.iplot(fig)
```
