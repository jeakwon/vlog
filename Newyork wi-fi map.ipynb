```python
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('nyc-wi-fi-hotspot-locations.csv')
```

```python
df.columns
```

```python
lat = list(df['Latitude'].values)
lon = list(df['Longitude'].values)

center_lat = df['Latitude'].quantile()
center_lon = df['Longitude'].quantile()
```

```python
mytoken = 'pk.eyJ1IjoiamVha3dvbiIsImEiOiJjamxhcGdpbjcxNHlxM3BvOHhxNWZ5dmh0In0.2oc7NG6uTtYd1mcqzBLNHQ'

location = go.Scattermapbox(
    lat = lat ,
    lon = lon ,
    marker = dict(
        size = 3
    )
)
data = [location]
layout = go.Layout(
    title = 'Newyork Wi-Fi Hotspot',
    mapbox = dict(
        center = dict(
            lat = center_lat,
            lon = center_lon,
        ),
        accesstoken = mytoken,
        style = 'dark',
        zoom = 8.5
    )
)
figure = go.Figure(data=data, layout=layout)
py.iplot(figure)
```

### Color with different Types of wi-fi

```python
df['Type'].unique()
```

```python
Fr = df[df['Type']=='Free']
Lf = df[df['Type']=='Limited Free']
Ps = df[df['Type']=='Partner Site']
```

```python
mytoken = 'pk.eyJ1IjoiamVha3dvbiIsImEiOiJjamxhcGdpbjcxNHlxM3BvOHhxNWZ5dmh0In0.2oc7NG6uTtYd1mcqzBLNHQ'

Fr_loc = go.Scattermapbox(
    lat = list(Fr['Latitude'].values) ,
    lon = list(Fr['Longitude'].values) ,
    name = 'Free',
    hovertext = Fr['Name'].values,
    marker = dict(
        size = 3,
        color = 'skyblue'
    )
)

Lf_loc = go.Scattermapbox(
    lat = list(Lf['Latitude'].values) ,
    lon = list(Lf['Longitude'].values) ,
    name = 'Limited Free',
    hovertext = Lf['Name'].values,
    marker = dict(
        size = 3,
        color = 'yellow'
    )
)

Ps_loc = go.Scattermapbox(
    lat = list(Ps['Latitude'].values) ,
    lon = list(Ps['Longitude'].values) ,
    name = 'Partner Site',
    hovertext = Ps['Name'].values,
    marker = dict(
        size = 3,
        color = 'red'
    )
)

data = [Fr_loc,Lf_loc,Ps_loc]
layout = go.Layout(
    title = 'Newyork Wi-Fi Hotspot',
    mapbox = dict(
        center = dict(
            lat = center_lat,
            lon = center_lon,
        ),
        accesstoken = mytoken,
        style = 'dark',
        zoom = 8.5
    )
)
figure = go.Figure(data=data, layout=layout)
py.iplot(figure)
```
